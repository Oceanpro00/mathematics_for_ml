```python
# Import Dependencies
import matplotlib.pyplot as plt
import random
```

```python
# Define Vector function
def vectorize(vector):
    
    x, y = 0, 0
    
    xs = [x]
    ys = [y]
    
    plt.figure()
    
    for i in range(0, len(vector), 2):
        dx, dy = vector[i], vector[i+1]
        
        quiver = plt.quiver(x, y, dx, dy, angles='xy', scale_units='xy', scale=1, color='blue')
        
        x += dx
        y += dy
        
        xs.append(x)
        ys.append(y)
    
    # define min and max values
    min_x, max_x = min(xs), max(xs)
    min_y, max_y = min(ys), max(ys)
    
    plt.axis('scaled')
    plt.xlim(min_x-3, max_x+3)
    plt.ylim(min_y-3, max_y+3)
    plt.grid()

    plt.title('Vector Plot')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
# vector list
flat_vector = []
```

```python
# Vectors we are adding up
vector_x = 2
vector_y = -2

# Addition or Subtraction
mode = 'add'

# scalar value
scalar = 5

for i in range(0,scalar):
    if mode != 'add':    
        flat_vector.append(vector_x*-1)
        flat_vector.append(vector_y*-1)
    else:
        flat_vector.append(vector_x)
        flat_vector.append(vector_y)

flat_vector

# Vectorize points
vectorize(flat_vector)
```

```python
# Define Vector pairs
vector_pairs = [(flat_vector[i], flat_vector[i+1]) for i in range(0,len(flat_vector), 2)]

# Randomize order
random.shuffle(vector_pairs)

flat_vector_check = [coord for pair in vector_pairs for coord in pair]

# See that order of addition does not affect result
vectorize(flat_vector_check)
```
