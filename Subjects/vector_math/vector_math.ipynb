```python
# Import Dependencies
import matplotlib.pyplot as plt
import random
```

### Basic flat Additive Vector Function

```python
# Define Vector function
def vectorize(vector, title = "Vector Plot"):
    
    x, y = 0, 0
    
    xs = [x]
    ys = [y]
    
    plt.figure()
    
    for i in range(0, len(vector), 2):
        dx, dy = vector[i], vector[i+1]
        
        quiver = plt.quiver(x, y, dx, dy, angles='xy', scale_units='xy', scale=1, color='blue')
        
        x += dx
        y += dy
        
        xs.append(x)
        ys.append(y)
    
    # define min and max values
    min_x, max_x = min(xs), max(xs)
    min_y, max_y = min(ys), max(ys)
    
    plt.axis('scaled')
    plt.xlim(min_x-3, max_x+3)
    plt.ylim(min_y-3, max_y+3)
    plt.grid()

    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

    # Return Final Position
    return x, y
    
```

#### Readability Function (Bold/ Subheadings)

```python
# Bolding Function
def bold(text):
    return f"\033[1;91m{text}\033[0m"

# Subheading Function
def subheading(text):
    print(f"\n\033[1m {text.upper()}\033[0m")
```

### Vector Addition Visualizer (2D)

This notebook helps you explore 2D vector behavior using **basic addition, scalar multiplication**, and **commutative properties**.

You can:
- Manually input vectors
- Adjust how many times to repeat them (scalar)
- Switch between **addition** and **subtraction**
- Shuffle the order to observe **commutativity**
- View a history of your vector operations

Each plot shows cumulative vector movement using tip-to-tail chaining.

> Built for intuition â€” no complex math libraries or vector classes needed!


```python
# vector list
flat_vector = []

# Vector Addition History
vector_history = []
```

```python
# Additive Vector Box
# Vectors we are adding up
vector_x = 0
vector_y = -1

# Addition or Subtraction
mode = 'add'

# scalar value
scalar = 5

for i in range(0,scalar):
    if mode != 'add':    
        flat_vector.append(vector_x*-1)
        flat_vector.append(vector_y*-1)
    else:
        flat_vector.append(vector_x)
        flat_vector.append(vector_y)

flat_vector

# Vectorize points
x, y = vectorize(flat_vector, "Ordered Vector Plot")

# History Summary Addition
summary = f"""Ordered Vector Addition (Mode: {bold(mode)}, Repeats: {bold(scalar)})
                Each vector added: [{bold(vector_x)}, {bold(vector_y)}]"""

vector_history.append(summary)

subheading("Vector Addition History")
for i, item in enumerate(vector_history,1):
    print(f"Addition #{i} - {item}")

subheading("Final Vector Position is")
print(f"The net deplacement vector position is [{bold(x)}, {bold(y)}]")
```

```python
# Define Vector pairs
vector_pairs = [(flat_vector[i], flat_vector[i+1]) for i in range(0,len(flat_vector), 2)]

# Randomize order
random.shuffle(vector_pairs)

flat_vector_check = [coord for pair in vector_pairs for coord in pair]

# See that order of addition does not affect result
vectorize(flat_vector_check, 'Shuffled Vector Plot')

# Explanations
subheading("Shuffled Vector Addition (Commutativity Check)")
print("The vectors are the same but appear in a random order.")
print("Notice that the endpoint remains the same, demonstrating commutativity.")

subheading("Final Vector Position is")
print(f"The net deplacement vector position is [{bold(x)}, {bold(y)}]")
```
